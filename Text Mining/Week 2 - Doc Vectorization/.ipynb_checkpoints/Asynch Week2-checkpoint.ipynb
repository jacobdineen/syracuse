```python
import pandas as pd
import sys
import glob, os
import numpy as np

path = os.getcwd() #Get Current working Directory
movies = os.listdir(path)[4] #Get path to movie review files
import nltk,csv,numpy  #import packages
from nltk import sent_tokenize, word_tokenize, pos_tag#import packages

import csv#import package

#read in csv
datafile = open(movies, 'r', newline = '') 
myreader = csv.reader(datafile, dialect='excel', delimiter='\t')

empty = [] #Create empty list
for i in myreader:
    empty.append(i) #append to empty list
    
import itertools
merged = list(itertools.chain(*empty)) #Compress list of lists down to a single list

tokens = [] #Create empty list
for i in merged:
    tokens.append(word_tokenize(i)) #Append tokens to new list
    
un_tokens = np.unique(tokens) #Check length of unique tokens


#Step to remove all punctuation
import re
clean = []

for i in merged:
    clean.append(str(re.sub(r'[^\w\s]','', i)))
    
clean[0] # Show a few examples 
```
